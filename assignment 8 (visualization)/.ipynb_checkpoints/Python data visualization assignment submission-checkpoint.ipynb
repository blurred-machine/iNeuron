# Python data visualization assignment submission

1) In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
df = datasets.load_iris()
X = df.data
y = df.target

fig = plt.figure(1, figsize=(10, 10))
ax = Axes3D(fig, elev=-150, azim=100)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.Set1, s=100)
ax.set_title("Iris 3D plot")
plt.show()
```
